```python
import pandas as pd
import nfl_data_py as nfl_data

seasons = range(2015, 2025)

# Import necessary data sets for prediction
team_data = nfl_data.import_seasonal_data(seasons) # Team data over season range
schedules = nfl_data.import_schedules(seasons) # Historical results over season range
team_desc = nfl_data.import_team_desc() # For properly joining datasets
weekly_trends = nfl_data.import_weekly_data(seasons) # weekly data to analyze weekly performances
rosters = nfl_data.import_seasonal_rosters(seasons) # For roster changes over the years
```

```python
# Show the shape of each dataset (rows, columns)
print("Team Dataset Shape: ", team_data.shape)
print("Schedule Dataset Shape: ", schedules.shape)
print("Team Stats Dataset Shape: ", team_desc.shape)
print("Weekly Trends Dataset Shape: ", weekly_trends.shape)
print("Roster Dataset Shape: ", rosters.shape)

# save the raw data into .csv files
import os

os.makedirs("../data/raw", exist_ok=True)

team_data.to_csv("../data/raw/team_data.csv", index=False)
schedules.to_csv("../data/raw/schedules.csv", index=False)
team_desc.to_csv("../data/raw/team_desc.csv", index=False)
weekly_trends.to_csv("../data/raw/weekly_trends.csv", index=False)
rosters.to_csv("../data/raw/rosters.csv", index=False)

print("Raw data saved!\n")
```
